```python
import pandas as pd
import seaborn as sns
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet('/Users/jiangwolin/IdeaProjects/scala-test/data/intermediateResults/trips')
df["bin"] = (df["distance"]/10).astype(int)
dist_count = df["bin"].value_counts().reset_index()
dist_count["count"] = dist_count["count"] / sum(dist_count["count"])
```

```python
sns.scatterplot(dist_count, x='bin', y='count')
plt.yscale('log')
plt.xscale('log')
plt.xlabel("distance")
plt.ylabel("probability")
plt.show()
```
